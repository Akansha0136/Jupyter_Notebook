```python
'''
This project involves visualizing a complex employee dataset to uncover insights related to employee experience, salary, 
performance, and productivity. Using various scatter plots, we analyze the relationships between different metrics such as years
of experience, salary, performance scores, and project completion to support data-driven decisions in human resource management.
'''
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("complex_dataset_with_random_names.csv")
df
```

```python
#To determine if there's a correlation between years of experience and salary, helping evaluate fairness in compensation.
plt.scatter(df['Experience_Years'], df['Salary'], c="blue")
plt.show()
```

```python
#To explore if employees who complete more projects also receive higher performance scores.
plt.scatter(df['Projects_Completed'], df['Performance_Score'], c="red")
plt.show()
```

```python
#To analyze whether age influences the number of overtime hours worked, possibly indicating generational work patterns.
plt.scatter(df['Age'], df['Overtime_Hours'], c="blue")
plt.show()
```

```python
#To examine how salaries have changed over employee joining dates, identifying trends in pay growth or inflation.
plt.plot(df['Joining_Date'],df['Salary'])
plt.legend(["Joining_Date", "Salary"])
plt.title("Linear graph", fontsize=25, color="green")
plt.ylabel('Salary')
plt.xlabel('Joining_Date')
plt.show()
```

```python
#To track how performance scores vary with employee tenure, suggesting whether performance improves or declines over time.
plt.plot(df['Joining_Date'],df['Performance_Score'])
plt.legend(["Joining_Date", "Performance_Score"])
plt.title("Linear graph", fontsize=25, color="green")
plt.ylabel('Performance_Score')
plt.xlabel('Joining_Date')
plt.show()
```

```python
#To understand how employee experience aligns with their joining date, validating tenure data accuracy.
plt.plot(df['Joining_Date'],df['Experience_Years'])
plt.legend(["Joining_Date", "Experience_Years"])
plt.title("Linear graph", fontsize=25, color="green")
plt.ylabel('Experience_Years')
plt.xlabel('Joining_Date')
plt.show()
```

```python
#To check whether promotions are aligned with productivity (in terms of project completion).
plt.plot(df['Last_Promotion_Date'],df['Projects_Completed'])
plt.legend(["Last_Promotion_Date", "Projects_Completed"])
plt.title("Linear graph", fontsize=25, color="green")
plt.ylabel('Projects_Completed')
plt.xlabel('Last_Promotion_Date')
plt.show()
```

```python
#To compare average or total salary distributions across departments, which can help in budget planning or fairness audits.
plt.bar(df['Department'], df['Salary'])
# Adding title to the plot
plt.title("Bar Chart")
# Adding label on the y-axis
plt.ylabel('Salary')
# Adding label on the x-axis
plt.xlabel('Department')
plt.show()
```

```python
#To assess departmental productivity by evaluating total projects completed per department.
plt.bar(df['Department'], df['Projects_Completed'])
# Adding title to the plot
plt.title("Bar Chart")
# Adding label on the y-axis
plt.ylabel('Projects_Completed')
# Adding label on the x-axis
plt.xlabel('Department')
plt.show()
```

```python
#To visualize the proportion of employees in each department.
department_counts = df["Department"].value_counts()

# Plotting the Donut Chart
plt.figure(figsize=(7, 7))
plt.pie(
    department_counts,
    labels=department_counts.index,
    autopct='%1.1f%%',  # Show percentage
    startangle=90,  # Start from the top
    wedgeprops={'width': 0.4},  # Width of the wedges to create the donut effect
    colors=plt.cm.tab10.colors  # Set a color palette
)

# Add a central circle for the "hole"
plt.gca().add_artist(plt.Circle((0, 0), 0.7, color='white'))

# Add title
plt.title("Employee Distribution by Department")
plt.show()
```
